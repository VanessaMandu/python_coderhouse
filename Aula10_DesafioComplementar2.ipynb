```python
'''Importacao do arquivo e bibliotecas'''

import pandas as pd

nintendo_df = pd.read_excel("best-selling_game_consoles.xlsx",sheet_name='consoles')
nintendo_df.head()
```

```python
'''Usando o arquivo "best-selling game consoles.xlsx" conte quantos consoles tem na base por companhias'''

nintendo_df.groupby('Company')['Type'].count().sort_values(ascending=False)
```

```python
'''Ordene o resultado anterior pela quantidade de consoles por companhia de forma decrescente'''
```

```python
'''Sobre o dataframe anterior, adicione a coluna "Market Value" presente na tabela do arquivo "company.txt"'''

mv_df = pd.read_csv("company.txt")
mv_df.head()
```

```python
novo_df = nintendo_df.merge(mv_df, on= 'Company', how= 'left')
novo_df.head()
```

```python
'''Qual o valor médio de Market Value por console?'''

novo_df.groupby('Console Name')['Market Value'].mean()
```
